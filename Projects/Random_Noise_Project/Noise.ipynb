### All Black

```python
image = np.zeros(resolution**2)
image = np.reshape(image, (resolution, resolution))

plt.figure(figsize = (5,5))
plt.imshow(image, cmap = 'gray')
plt.show()
```

### Gradient

```python
image = np.arange(0, resolution**2)
image = np.reshape(image, (resolution, resolution))

plt.figure(figsize = (5,5))
plt.imshow(image, cmap = 'gray')
plt.show()
```

```python
image = np.zeros(resolution**2)
image = np.reshape(image, (resolution, resolution))

for i in range(0, resolution):
    for j in range(0, resolution):
        if (i % 2 == 0 and j % 2 == 0) or (not i % 2 == 0 and not j % 2 == 0):
            image[i][j] = 255

plt.figure(figsize = (16, 16))
plt.imshow(image, cmap = 'gray')
plt.show()
```

### Completely Random

```python
image = np.random.rand(resolution**2)
image = np.reshape(image, (resolution, resolution))

plt.figure(figsize = (5, 5))
plt.imshow(image, cmap = 'gray')
plt.show()
```
